## Condicional if

```python
# Condidicional if (Se)
if 5 > 2:
  print ("A sentençã é verdadeira!")
```

```python
# Condidcional if...else
if 5 < 2:
  print ("A sentençã é verdadeira!")
else:
  print ("A sentençã é falsa!")
```

```python
# Condicional if...else com variável
dia = "Quarta-feira"
if dia == "Sexta-feira":
  print("Hoje fará sol!")
else:
  print("Hoje vai chover!")
```
